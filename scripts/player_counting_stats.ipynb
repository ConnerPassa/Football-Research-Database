# Script for Retrieving Basic Player Counting Stats

```python
import pandas as pd
import numpy as np
import nfl_data_py as nfl
import matplotlib as plt
import seaborn as sns
from functools import reduce
```

You need to first declare the seasons you want.
Second import the seasonal data.
Third you need to import rosters because you'll need it for a merge later.

```python
seasons = range(2010,2024+1)
basic_stats = nfl.import_seasonal_data(seasons, 'REG')

rosters = nfl.import_seasonal_rosters(seasons, ['season', 'team', 'position', 'player_name', 'player_id'])
```

Merge the basic_stats with rosters so that you can get the player names.

```python
player_counting_stats = basic_stats.merge(rosters[['season', 'player_id', 'player_name']],
                                          how = 'left',
                                          on = ['season', 'player_id'])
```

```python
# Move player_name to the first column
cols = ['player_name'] + [col for col in player_counting_stats.columns if col != 'player_name']
player_counting_stats = player_counting_stats[cols]

player_counting_stats
```

```python
player_counting_stats.to_csv('player_counting_stats.csv', index = False)
```
